```python
import pandas as pd
import os

csv_folder = r'./adventureworks_csv'
csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

dataframes = {}

encodings_to_try = ['utf-8', 'utf-16', 'ISO-8859-1']

for file in csv_files:
    df_name = file.replace('.csv', '')
    file_path = os.path.join(csv_folder, file)

    for enc in encodings_to_try:
        try:
            # ✅ Using tab as the separator
            df = pd.read_csv(file_path, encoding=enc, sep='\t', on_bad_lines='skip', low_memory=False)
            dataframes[df_name] = df
            print(f"✅ Loaded {df_name} using encoding {enc}: {df.shape[0]} rows, {df.shape[1]} columns")
            break
        except Exception as e:
            last_error = e
    else:
        print(f"❌ Failed to load {df_name} with all encodings. Last error: {last_error}")
```
